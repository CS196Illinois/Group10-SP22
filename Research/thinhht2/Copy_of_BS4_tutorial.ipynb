```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://courses.illinois.edu/schedule/2022/spring")
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
list(soup.children)[2]
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup.find_all("a")
```

```python
majors_ = [] 
for row in soup.findAll('tr')[1:]:
    col = row.findAll('td')
    majors_.append(col[0].getText())
print(majors_)
```

```python
cleaned_ = []
for item in majors_:
    cleaned_.append(item.strip().replace("\n", ""))
```

```python
print(cleaned_)
```

```python
for item in cleaned_:
  major = "https://courses.illinois.edu/schedule/2022/spring/" + item
  print(major)
```
