```python
#Function definition
def base_to_base(baseNumber):
    baseList = []
    while baseNumber > 0:
        binaryValue = baseNumber % 2
        baseList.append(binaryValue)
        baseNumber = baseNumber / 2
    baseList2 = []
    for i in reversed(range(len(baseList))) :
        baseList2.append(i)
    baseNumberDict = {
        baseNumber : baseList2
    }
    return baseList
```
